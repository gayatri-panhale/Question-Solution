<h3>Ex.2.check the goodness of fit for the following data mapping number of hours a student studies v/s the expected number of hours at 5% los abserved data(in hours) - 8 6 10 7 8 11 9 expected data(in hours)- 9 8 11 8 10 7 6

```python

import numpy as np
from scipy.stats import chisquare


observed = np.array([8, 6, 10, 7, 8, 11, 9])


expected = np.array([9, 8, 11, 8, 10, 7, 6])


chi_square, p_value = chisquare(f_obs=observed, f_exp=expected)


print("Chi-Square Value =", chi_square)


print("P-Value =", p_value)


if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")

# final conclusion
print("Conclusion: Data shows good fit")
```
