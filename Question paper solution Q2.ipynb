<h3>Ex.2.what is the probability of 2 heads in 6 coin tosses?fit the binomial distribution for the coin tosses and ploy it.

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom


n = 6


p = 0.5


x = np.arange(0, 7)


y = binom.pmf(x, n, p)


for i in range(len(x)):
    print("P(X =", x[i], ") =", y[i])


print("\nProbability of exactly 2 heads =", y[2])


plt.bar(x, y)


plt.xlabel("Number of Heads")


plt.ylabel("Probability")


plt.title("Binomial Distribution for 6 Coin Tosses")


plt.show()
```
